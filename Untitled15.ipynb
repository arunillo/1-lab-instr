```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Загрузка данных
tickers = ["KO", "CLF", "UMC"]
data = yf.download(tickers, start="2019-01-01", end="2019-12-31")['Close']
returns = data.pct_change().dropna()

# 2. График стоимости бумаги с минимальной доходностью
min_return_ticker = returns.mean().idxmin()
plt.figure(figsize=(10, 4))
plt.plot(data[min_return_ticker])
plt.title(f'{min_return_ticker} - Динамика стоимости')
plt.ylabel('Цена ($)')
plt.grid(True)
plt.show()

# 3. Boxplot для бумаги с максимальной доходностью
max_return_ticker = returns.mean().idxmax()
plt.figure(figsize=(8, 5))
plt.boxplot(returns[max_return_ticker])
plt.title(f'{max_return_ticker} - Диаграмма размаха доходностей')
plt.ylabel('Доходность')
plt.grid(True)
plt.show()

# 4. ECDF для бумаги с минимальной дисперсией
min_var_ticker = returns.var().idxmin()
x = np.sort(returns[min_var_ticker])
y = np.arange(1, len(x)+1) / len(x)
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b.-')
plt.title(f'{min_var_ticker} - Эмпирическая функция распределения')
plt.xlabel('Доходность')
plt.ylabel('ECDF')
plt.grid(True)
plt.show()

# 5. Scatter plot для пары с минимальной корреляцией
corr_matrix = returns.corr()
pairs = [(i, j, corr_matrix.loc[i, j]) for i in tickers for j in tickers if i < j]
min_corr_pair = min(pairs, key=lambda x: abs(x[2]))
plt.figure(figsize=(8, 5))
plt.scatter(data[min_corr_pair[0]], data[min_corr_pair[1]], alpha=0.5)
plt.xlabel(min_corr_pair[0])
plt.ylabel(min_corr_pair[1])
plt.title(f'Диаграмма рассеяния стоимостей\nКорреляция: {min_corr_pair[2]:.3f}')
plt.grid(True)
plt.show()

# Результаты
print(f"Min доходность: {min_return_ticker} ({returns.mean()[min_return_ticker]:.6f})")
print(f"Max доходность: {max_return_ticker} ({returns.mean()[max_return_ticker]:.6f})")
print(f"Min дисперсия: {min_var_ticker} ({returns.var()[min_var_ticker]:.8f})")
print(f"Min корреляция: {min_corr_pair[0]}-{min_corr_pair[1]} ({min_corr_pair[2]:.3f})")
```
